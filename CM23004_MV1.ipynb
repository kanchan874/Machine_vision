```python
 import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

filename = next(iter(uploaded))
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

equalized = cv2.equalizeHist(gray)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalized")

plt.show()
```

```python
bright = cv2.convertScaleAbs(img, alpha=1.1, beta=20)

plt.imshow(bright)
plt.title("Brightness Enhanced")
plt.axis('off')
plt.show()
```

```python
median = cv2.medianBlur(bright, 5)

plt.imshow(median)
plt.title("After Median Filtering")
plt.axis('off')
plt.show()
```

```python
gaussian = cv2.GaussianBlur(median, (5,5), 0)

plt.imshow(gaussian)
plt.title("After Gaussian Filtering")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15,8))

plt.subplot(2,3,1)
plt.imshow(img)
plt.title("Original")
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalization")
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(bright)
plt.title("Brightness Enhanced")
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(median)
plt.title("Median Filter")
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(gaussian)
plt.title("Gaussian Filter")
plt.axis('off')

plt.tight_layout()
plt.show()
```
